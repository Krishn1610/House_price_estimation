```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("House.csv")
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
stats = data["Price"].describe().reset_index()
stats["Price"] =  round(stats["Price"],2)
stats
```

```python
data.isna().sum().sum()
```

```python
data.duplicated().sum()
```

```python
data.dropna(inplace =True)
data.drop_duplicates(inplace= True)
```

```python
data.head()
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind = "bar")
plt.title("Condition of the house vs Price")
plt.ylabel("Mean Price")
plt.show()
```

```python
data.columns
```

```python
data[["number of bedrooms", "number of bathrooms"]]
```

```python
X = data[["number of bedrooms", "number of bathrooms", "living area", "condition of the house", "Number of schools nearby" ]]
y = data[["Price"]]
```

```python
y
```

```python
data.shape[0]*0.8
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_model = DecisionTreeRegressor()
```

```python
param_grid = {
    'criterion': ['squared_error', 'absolute_error'],  # Fewer options
    'max_depth': [10, 20],  # Reduce the number of depth values
    'min_samples_split': [2, 5],  # Add relevant parameters but limit the range
}

grid_tree = GridSearchCV(estimator=tree_model, param_grid=param_grid, cv=5)
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [ 5, 10, 15],
    "n_estimators" : [ 2, 3, 4, 5, 6, 7, 8, 9, 10]
}
```

```python
gridrfr = GridSearchCV( rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, y_train)
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(X_test)
```

```python
mean_absolute_error( y_test, rfrpredictions)
```
